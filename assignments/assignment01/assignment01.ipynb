## Compare "Sports" and "Comady", how many days that there are more total daily views of VDO in "Sports" category than in "Comady" category?
# ANS: 49

```python
# sport 17 comady 23
ta
```

```python
sport = terra[terra.category_id==17].groupby('trending_date')['views'].sum().to_frame()
comady = terra[terra.category_id==23].groupby('trending_date')['views'].sum().to_frame()
```

```python
j=sport.join(comady, lsuffix='_sport', rsuffix='_comady')
j.fillna(0)
len(j[j.views_sport > j.views_comady])
```
